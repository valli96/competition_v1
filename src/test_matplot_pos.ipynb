```python
from tkinter import *
from tkinter.ttk import Progressbar
import time


def step():
    for i in range(5):
        ws.update_idletasks()
        pb['value'] += 20
        time.sleep(1)
        txt['text']=pb['value'],'%'

ws = Tk()
ws.title('PythonGuides')
ws.geometry('200x150')
ws.config(bg='#345')


pb = Progressbar(
    ws,
    orient = HORIZONTAL,
    length = 100,
    mode = 'determinate'
    )

pb.place(x=40, y=20)

txt = Label(
    ws,
    text = '0%',
    bg = '#345',
    fg = '#fff'

)

txt.place(x=150 ,y=20 )

Button(
    ws,
    text='Start',
    command=step
).place(x=40, y=50)

ws.mainloop()
```

```python
import tkinter as tk
import time

state = 1

def start_indicators():
    global state
    """
    This function updates the widgets based on your list
    :return: None
    """

    label_text = "Current value: {}".format(state)  # Set the text of Label
    if state == 0:
        print("Value: {} - Danger!".format(state))
        my_canvas.itemconfig(my_oval, fill="red")  # Fill the circle with RED
        label_var.set(label_text)  # Updating the label
    elif state == 1:
        print("Value: {} - Warning!".format(state))
        my_canvas.itemconfig(my_oval2, fill="yellow")  # Fill the circle with YELLOW
        label_var.set(label_text)  # Updating the label
    elif state == 2:
        print("Value: {} - Safe!".format(state))
        my_canvas.itemconfig(my_oval3, fill="green")  # Fill the circle with GREED
        label_var.set(label_text)  # Updating the label
    root.update()  # Update the complete GUI.
    print("FINISH")


root = tk.Tk()  # Set Tk instance
my_canvas = tk.Canvas(root, width=200, height=600)  # Create 200x200 Canvas widget
my_canvas.pack()

my_oval = my_canvas.create_oval(50, 50, 100, 100)  # Create a circle on the Canvas
my_oval2 = my_canvas.create_oval(50, 100, 100, 150)  # Create a circle on the Canvas
my_oval3 = my_canvas.create_oval(50, 150, 100, 200)  # Create a circle on the Canvas

label_var = tk.StringVar()  # Set the string variable for Label widget

my_label = tk.Label(root, textvariable=label_var)  # Set the LAbel widget
my_label.pack()

my_button = tk.Button(root, text="START", command=start_indicators)  # Set a "Start button". The "start_indicators" function is a call-back.
my_button.pack()

root.mainloop()
```
